```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
```

```python
df = pd.read_csv('Student_performance_data _.csv')
df
```

```python
df_gpa = df.loc[:, 'GPA']
df_gpa
```

Teste T-Student:

HipÃ³tese nula (H0): a mÃ©dia das notas (GPA) Ã© igual a 2.0

HipÃ³tese alternativa (H1): a mÃ©dia das notas (GPA) Ã© diferente de 2.0

```python
h0 = 2
h1 = 'two-sided'
media_gpa = df_gpa.mean()
print('MÃ©dia populacional:', h0)
print('MÃ©dia amostral:', media_gpa)
t_statistic, p_value = stats.ttest_1samp(df_gpa, popmean= h0)

print('EstatÃ­stica t:', t_statistic)
print('Valor p:', p_value)
```

```python
if p_value <= 0.05:
    print(f"Rejeitamos a hipÃ³tese nula, pois a mÃ©dia do GPA dos alunos Ã© significativamente diferente de {h0}.")
else:
    print(f"NÃ£o rejeitamos a hipÃ³tese nula, pois a mÃ©dia do GPA dos alunos Ã© diferente de {h0}.")
```

```python
plt.hist(df_gpa, bins=10, edgecolor='black')
plt.xlabel('GPA')
plt.ylabel('FrequÃªncia')
plt.title('DistribuiÃ§Ã£o de FrequÃªncias')

plt.axvline(h0, color='red', linestyle='--', label='MÃ©dia Populacional (2.0)')
plt.axvline(df_gpa.mean(), color='blue', linestyle='-', label='MÃ©dia Amostral')

plt.legend()
plt.show()
```
